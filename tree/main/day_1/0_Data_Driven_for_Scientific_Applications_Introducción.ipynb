# **Guardar y compartir una Notebook en Google Colaboratory**

Luego de finalizar un análisis o proyecto en Colab, es probable que quiera guardar su cuaderno de R para poder acceder a él más tarde o compartirlo con otros colaboradores, ya sea de forma pública o privada.

Para guardar su cuaderno R en Colab, puede hacer clic en "Archivo" en la esquina superior izquierda y seleccionar "Guardar" o usar el método abreviado de teclado "Ctrl+S". Colab guardará su computadora portátil en su cuenta de Google Drive, donde podrá acceder a ella más tarde.

Haciendo clip en el botón compartir tiene un menú emergente similar a otras documentos o proyectos en la nube de google, allí puedes generar un enlace para compartirlo. también puedes compartir con colaboradores específicos por medio de un correo electrónico.
